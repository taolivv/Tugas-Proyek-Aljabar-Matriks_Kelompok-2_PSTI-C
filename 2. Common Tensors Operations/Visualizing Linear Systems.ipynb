Buat visualisasi dari persamaan berikut:
$$ y = 2x+1  $$ $$ 3x-y = 5 $$


Ubah persamaan kedua dalam bentuk persamaan y
$$ 3x - y = 5 $$
$$ - y = 5 - 3x $$
$$  y = 3x + 5 $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 5, 200)
y1 = 2 * x + 1
y2 = 3 * x + 5
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=-4, color='purple', linestyle='--')
plt.axhline(y=-7, color='purple', linestyle='--')
plt.grid()
plt.show()
```
