Nomor 1


```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
import numpy as np

X = np.array([[10, 20, 30], [15, 30, 45]])
print("Matrix X:\n", X)
```

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
import torch as pt

X_pt = pt.tensor([[10, 20, 30], [15, 30, 45]])
print("Matrix X_pt:\n", X_pt)
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
import tensorflow as tf

X_tf = tf.Variable([[10, 20, 30], [15, 30, 45]])
print("Matrix X_tf:\n", X_tf)
```

Nomor 2

```python
# lakukan transpose pada matrix X
X_trans = X.T
print("Transpose matrix X:\n", X_trans)

# lakukan transpose pada matrix X_pt
X_pt_trans = X_pt.T
print("\nTranspose matrix X_pt:\n", X_pt_trans)

# lakukan transpose pada matrix X_tf
X_tf_trans = tf.transpose(X_tf)
print("\nTranspose matrix X_tf:\n", X_tf_trans)
```
