```python
# buat matrix A dengan ukuran 2x3 dan vector b berukuran 3x1 menggunakan numpy
import numpy as np
A = np.array([[10, 20, 30], [15, 30, 45]])
print("Matrix A: \n",A)
b = np.array([[1], [2], [3]])
print("\nVector b: \n",b)
```

```python
# buat matrix A_pt dengan ukuran 2x3 dan vector b_pt berukuran 3x1 menggunakan pytorch
import torch as pt
A_pt = pt.tensor([[10, 20, 30], [15, 30, 45]])
print("Matrix A: \n",A_pt)
b_pt = pt.tensor([[1], [2], [3]])
print("\nVector b: \n",b_pt)
```

```python
# buat matrix A_tf dengan ukuran 2x3 dan vector b_tf berukuran 3x1 menggunakan tensorflow
import tensorflow as tf
A_tf = tf.Variable([[10, 20, 30], [15, 30, 45]])
print("Matrix A: \n",A_tf)
b_tf = tf.Variable([[1], [2], [3]])
print("\nVector b: \n",b_tf)
```
