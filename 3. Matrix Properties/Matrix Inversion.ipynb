```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# buat matrix X ukuran 2x2 menggunakan numpy
X = np.array([[10, 12], [15, 20]])
print("Matrix X:\n", X)
```

```python
# buat invers dari matrix X dan beri nama Xinv menggunakan numpy
Xinv = np.linalg.inv(X)
print("Invers dari matrix X:\n", Xinv)
```

```python
# kalikan X dengan Xinv menggunakan numpy
print("Hasil perkalian matrix X dan Xinv:\n", np.dot(X, Xinv))
```

Persamaan linear:
$$ 4b + 2c = 4 $$
$$ -5b - 3c = -7 $$

Ubah persamaan kesatu dalam bentuk persamaan c:
$$ 4b + 2c = 4 $$
$$ 2b + c = 2 $$
$$ c = 2 - 2b $$

Ubah persamaan kedua dalam bentuk persamaan c:
$$ -5b - 3c = -7 $$
$$ -3c = -7 + 5b $$
$$ c = \frac{-7 + 5b}{-3} = \frac{7 - 5b}{3} $$

```python
#  dengan numpy
import numpy as np

A = np.array([[4, 2], [-5, -3]])
B = np.array([[4], [-7]])
C = np.linalg.solve(A, B)
print("Solusi [b, c] dari persamaan diatas:\n", C)

print("\nGrafiknya:")
b = np.linspace(-10, 10, 1000)
c1 = 2 - 2*b
c2 = (7-5*b)/3

fig, ax = plt.subplots()
plt.xlabel('b', c='darkorange')
plt.ylabel('c', c='brown')

plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

ax.set_xlim([-2, 3])
ax.set_ylim([-1, 5])
ax.plot(b, c1, c='purple')
ax.plot(b, c2, c='purple')
plt.axvline(x=-1, color='green', linestyle='--')
_ = plt.axhline(y=4, color='green', linestyle='--')
```

```python
# dengan pytorch
import torch as pt

A = pt.tensor([[4, 2], [-5, -3]], dtype=pt.float32)
B = pt.tensor([[4], [-7]], dtype=pt.float32)
C = pt.linalg.solve(A, B)
print("Solusi [b, c] dari persamaan diatas:\n", C)
```

```python
# dengan tensorflow
import tensorflow as tf

A = tf.Variable([[4, 2], [-5, -3]], dtype=tf.float64)
B = tf.Variable([[4], [-7]], dtype=tf.float64)
C = tf.linalg.solve(A, B)
print("Solusi [b, c] dari persamaan diatas:\n", C)
```
