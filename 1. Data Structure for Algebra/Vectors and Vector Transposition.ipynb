Nomor 1

```python
import numpy as np
```

```python
# buat vector x dengan ukuran 3x1
x = np.array([[1], [2], [3]])
print("Vektor x:\n", x)
```

```python
# hitung panjang vector x
print("Panjang vektor x:", len(x))
```

```python
# tampilkan ukuran vector x
print("Ukuran vektor x:", x.shape)
```

```python
# tampilkan tipe data dari vector x
print("Tipe data vektor x:", type(x))
```

```python
# akses elemen pertama dari vector x
print("Elemen pertama vektor x:", x[0])
```

```python
# tampilkan tipe data elemen pertama dari vector x
print("Tipe data elemen pertama vektor x:", type(x[0]))
```

Nomor 2

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t
x_t = x.T
print("vektor x_t:\n", x_t)
```

```python
# tampilkan ukuran dari vector x_t
print("ukuran vektor x_t:", x_t.shape)
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
y = np.array([[12, 14, 16]])
y_t = y.T
print("vektor y_t:\n", y_t)
```

```python
# tampilkan ukuran dari vector y_t
print("ukuran vektor y_t:", y_t.shape)
```

Lakukan soal nomor 2 menggunakan pytorch

```python
import torch
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_pt
x_pt = torch.tensor([[1], [2], [3]])
x_pt = x.T
print("vektor x_pt:\n", x_pt)
```

```python
# tampilkan ukuran dari vector x_pt
print("ukuran vektor x_pt:", x_pt.shape)
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_pt
y_pt = torch.tensor(y)
y_pt = y_pt.T
print("vektor y_pt:\n", y_pt)
```

```python
# tampilkan ukuran dari vector y_pt
print("ukuran vektor y_pt:", y_pt.shape)
```

Lakukan soal nomor 2 menggunakan tensorflow

```python
import tensorflow as tf
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_tf
x_tf = tf.constant([[1],
                   [2],
                   [3]])
x_tf = x.T
print("vektor x_tf:\n", x_tf)
```

```python
# tampilkan ukuran dari vector x_tf
print("ukuran vektor x_tf:", x_tf.shape)
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_tf
y_tf = tf.Variable([12, 14, 16])
y_tf = y.T
print("vektor y_tf:\n", y_tf)
```

```python
# tampilkan ukuran dari vector y_tf
print("ukuran vektor y_tf:", y_tf.shape)
```
