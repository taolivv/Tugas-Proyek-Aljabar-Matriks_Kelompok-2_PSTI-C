```python
import numpy as np
```

```python
# buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
i = np.array([1, 0])
print("Vector i:", i)
j = np.array([0, 1])
print("Vector j:", j)
```

```python
# buktikan bahwa vector i dan vector j itu orthogonal
bukti1 = np.dot(i, j)

if bukti1 == 0:
    print("i dan j adalah orthogonal, karena syarat dua vektor disebut orthogonal dot product kedua vektor harus = ", bukti1)
else:
    print("i dan j bukan orthogonal karena hasil akhirnya", bukti1)
```

```python
# Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1, 2])
w = np.array([2, -1])

bukti2 = np.dot(v, w)

if bukti2 == 0:
    print("v dan w juga orthogonal, karena hasil akhirnya:", bukti2)
else:
    print("v dan w bukan orthogonal karena hasil akhirnya:", bukti2)
```
