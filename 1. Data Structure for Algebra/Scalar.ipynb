Nomor 2, Lakukan tahapan pada soal 1 dengan menggunakan tensorflow

```python
import tensorflow as tf
```

```python
# buat variabel skalar x_tf dan simpan suatu nilai skalar
x_tf = tf.Variable(100, dtype=tf.int16)
print("Nilai x_tf:",x_tf)
```

```python
# tampilkan bentuk dan tipe data dari x_tf
print("Bentuk data x_tf:",x_tf.shape)
print("Tipe data x_tf:",type(x_tf))
```

```python
# buat variabel skalar y_tf dan simpan suatu nilai skalar
y_tf = tf.Variable(50, dtype=tf.int16)
print("Nilai y_tf:",y_tf)
```

```python
# tampilkan bentuk dan tipe data dari y_tf
print("Bentuk data y_tf:",y_tf.shape)
print("Tipe data y_tf:",type(y_tf))
```

```python
# jumlahkan x_tf + y_tf
tf_sum1 = x_tf + y_tf
print("Hasil dari x_tf + y_tf:",tf_sum1)

tf_sum2 = tf.add(x_tf, y_tf)
print("Hasil dari x_tf + y_tf:",tf_sum2)

print("Hasil dari x_tf + y_tf:",tf_sum1.numpy())
```

```python
# buat variabel bernama tf_float dan simpan suatu nilai desimal
tf_float = tf.Variable(0.25, dtype=tf.float16)
print("Nilai tf_float:",tf_float)
```

```python
# jumlahkan tf_float dengan y
sum_float = tf_float + y
print("Hasil dari tf_float + y:",sum_float)
```

```python
# tampilkan bentuk dan tipe data dari tf_float
print("Bentuk data tf_float:",tf_float.shape)
print("Tipe data tf_float:",type(tf_float))
```
