Nomor 1

```python
import numpy as np
```

```python
# buat matrix X berukuran 2x3 menggunakan numpy
X = np.array([[2, 4, 6], [3, 5, 7]])
print("Matrix X:\n", (X))
```

```python
# tampilkan ukuran dari matrix X
print("Ukuran matrix X:", X.shape)
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
print("Elemen kolom pertama dari matrix X:", X[:,0])
```

```python
#tampilkan seluruh elemen dari baris kedua dari matrix X
print("Elemen baris kedua dari matrix X:", X[1,:])
```

```python
import tensorflow as tf
```

```python
# buat matrix X berukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[2, 4, 6], [3, 5, 7]])
print("Matrix X_tf:\n", X_tf)
```

```python
#tampilkan seluruh elemen dari baris kedua dari matrix X_tf
print("Elemen baris kedua dari matrix X_tf:", X_tf[1,:])
```
